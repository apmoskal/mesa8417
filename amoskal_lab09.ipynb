```python
# prompt: make a altair bar chart of allocation subcategory and gift amount and send to gradio

def create_bar_chart(dataframe):
    chart = alt.Chart(dataframe).mark_bar().encode(
        x='Allocation Subcategory',
        y='Gift Amount'
    ).properties(
        title='Gift Amount by Allocation Subcategory'
    )
    return chart.to_json()

def gradio_interface(chart_json):
    return gr.JSON(chart_json)

iface = gr.Interface(fn=gradio_interface, inputs=gr.JSON(create_bar_chart(df)), outputs="json")
iface.launch(debug=True)
```

```python
from vega_datasets import data
```

```python
# Load dataset
cars = data.cars()


# Function to generate Altair scatter plot
def create_scatterplot():
    chart = alt.Chart(cars).mark_circle(size=60).encode(
        x="Horsepower",
        y="Miles_per_Gallon",
        color="Origin",
        tooltip=["Horsepower", "Miles_per_Gallon", "Origin"]
    )
    return chart



# Create Gradio Interface
demo = gr.Interface(
    fn=create_scatterplot,
    inputs=None,
    outputs=gr.Plot(),
    title="Altair Scatterplot in Gradio",
    description="This scatterplot shows the relationship between Horsepower and MPG in cars."
)


# Launch in notebook mode
demo.launch()
```
